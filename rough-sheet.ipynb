```python
def concentric_squre(matrix_number):
    matrix_dimension = 2 * matrix_number - 1
    print(f"matrix_dimension: {matrix_dimension} x {matrix_dimension}")

    for row in range(matrix_dimension):
        line = []
        for col in range(matrix_dimension):
            distance = [row, matrix_dimension - 1 - row , col, matrix_dimension - 1 - col]
            layer =  min(distance)
            value = matrix_number - layer
            line.append(str(value))
        print(' '.join(line))


concentric_squre(4)
```
