```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for idx_01 in range(len(nums)):
            for idx_02 in range(idx_01 + 1 , len(nums)):
                if nums[idx_01] + nums[idx_02] == target:
                    return [idx_01, idx_02]

sol = Solution()
print(sol.twoSum([2, 7, 11, 15], 18))
```

```python
import sys
from typing import List, Any, Set

def get_deep_size(obj: Any, seen: Set[int] = None) -> int:
    if seen is None:
        seen = set()

    obj_id = id(obj)
    if obj_id in seen:
        return 0
    seen.add(obj_id)

    size = sys.getsizeof(obj)

    if isinstance(obj, dict):
        size += sum(get_deep_size(k, seen) + get_deep_size(v, seen) for k, v in obj.items())
    elif isinstance(obj, (list, tuple, set, frozenset)):
        size += sum(get_deep_size(item, seen) for item in obj)

    return size

def format_memory(num_bytes: int) -> str:
    mb = num_bytes / 1_000_000
    mib = num_bytes / 1_048_576
    return f"{num_bytes} bytes ({mb:.6f} MB, {mib:.6f} MiB)"

def print_object_memory(name: str, obj: Any) -> None:
    print(f"{name} memory: {format_memory(get_deep_size(obj))}")

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hash_map_val = {}
        for idx, value in enumerate(nums):
            print(f"idx: {idx}, value: {value}")

            complement = target - value
            print(f"complement: {complement}")

            if complement in hash_map_val:
                print(f"hash_map_val memory: {format_memory(get_deep_size(hash_map_val))}")
                return [hash_map_val[complement], idx]

            hash_map_val[value] = idx

        print(f"hash_map_val memory: {format_memory(get_deep_size(hash_map_val))}")
        return []

nums_input = [2, 11, 17, 7]
target_input = 18
sol = Solution()
result = sol.twoSum(nums_input, target_input)
print(result)
print_object_memory("nums_input", nums_input)
```

```python
# Example 1: Using Profiler as Decorator
@profile_function
def two_sum_approach_1(nums, target):
    """Brute force O(n^2) approach"""
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []

@profile_function
def two_sum_approach_2(nums, target):
    """Hash map O(n) approach"""
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

# Test both approaches
test_nums = [2, 7, 11, 15]
test_target = 18

print("Approach 1 (Brute Force):")
result1 = two_sum_approach_1(test_nums, test_target)

print("Approach 2 (Hash Map):")
result2 = two_sum_approach_2(test_nums, test_target)
```
