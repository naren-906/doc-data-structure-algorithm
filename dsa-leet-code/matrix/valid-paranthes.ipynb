# Valid Parentheses

## Problem Statement
Given a string `s` containing just the characters `'('`, `')'`, `'['`, `']'`, `'{'` and `'}'`, determine if the input string is valid.

An input string is valid if:
- Every open bracket is closed by the same type of bracket
- The brackets are closed in the correct order

## Examples

### Example 1
- **Input:** `s = "{[]}"`
- **Output:** `true`

## Constraints
- `1 ≤ s.length ≤ 10⁴`
- `s` consists of parentheses only: `()[]{}` 

## Solution


```python
# Brute force approach:

def isValid(s):
    while "()" in s or "{}" in s or "[]" in s:
        print("s: ", s)
        s = s.replace("()", "").replace("{}", "").replace("[]", "")
    return s == ""

print(isValid("([{}])"))
print('------')
print(isValid("([{]})"))
```

```python
# Stack approach

def isValid(s):
    stack = []
    pairs = {
        ")" : "(",
        "}" : "{",
        "]" : "[",
    }
    
    for paranth in s:
        if paranth in "({[":
            stack.append(paranth)
        else:
            if not stack:
                print('--- 1 ---')
                return False
            
            # print(stack)
            # print(stack[-1])
            # print(paranth)
            # print(pairs[paranth])
            
            if stack[-1] != pairs[paranth]:
                print('--- 2 ---')
                return False
            
            stack.pop()
            
    print(stack)
    return not stack

print(isValid("([{}])"))
print('--- break ---')
print(isValid("([{]})"))
print('--- break ---')
print(isValid(""))
```

```python
# 4. Recursive Solution

def isValid(s):
    try:
        print("step 1: input string -> ", s)
        
        new_string = s.replace("()", "").replace("{}", "").replace("[]", "")
        print("step 2: string replacement -> ", new_string)
        
        if new_string == "":
            print("step 3: if string empty return -> ", new_string)
            return True
        
        print("step 4: recrusive function call -> ", new_string)
        return isValid(new_string)
    
    except Exception as e:
        print("step 5: max recrusive reached", new_string)
        return False


# print(isValid("([{}])"))
# print('--- break ---')
print(isValid("([{]})"))
print('--- break ---')
print(isValid(""))
```
