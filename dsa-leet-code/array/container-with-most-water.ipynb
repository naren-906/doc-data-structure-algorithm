```python
# 11. Container With Most Water (Brut-Force)
from typing import List

class Solution:
    def mostWater(self, height: List[int]) -> int:
        """_summary_
            - [1, 8, 6, 2, 5, 4, 8, 3, 7]
            - Brut-force approach:
                - loop every element in the element
                - Hint 1: find width of the frst and last value
                    width = idx_right - idx_left
                
                - Hint 2: find height for contained water(smaller of the two heights)
                    height = min(idx_right, idx_left)
                
                - Hint 3: find area of water contined
                    area = width * height
 
        """
        most_water_area, indices_most_water_area = 0, (0,0)
        print(f"Input: {height}")
        print('-------------------------------------')
        for idx_01 in range(len(height)-1):
            for idx_02 in range(idx_01+1, len(height)):
                container_width = idx_02 - idx_01
                # print(f"1. container_width: {idx_02} - {idx_01} = {container_width}")
                
                container_height = min(height[idx_01], height[idx_02])
                # print(f"2. container_height: min({height[idx_01]}, {height[idx_02]}) = {container_height}")
                
                area = container_width * container_height
                # print(f"3. area: {container_width} * {container_height} = {area}")
                # print('-------------------------------------')
                if area > most_water_area:
                    most_water_area = area
                    indices_most_water_area = (idx_01, idx_02)

        print(f"output: most_water_area -> {most_water_area}")
        print(f"output: indices_most_water_area -> {indices_most_water_area}")
        return most_water_area
    
input_height = [1, 8, 6, 2, 5, 4, 8, 3, 7]
obj_solution = Solution()
water_area = obj_solution.mostWater(input_height)
```
