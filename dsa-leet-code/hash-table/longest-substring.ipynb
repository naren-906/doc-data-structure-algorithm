```python
class Solution:
    def lengthOfLongestSubstring(s: str) -> int:
        max_length = 0
        for i_number in range(len(s)):
            seen = set()
            
            for j_number in range(len(s)):
                if s[j_number] in seen:
                    break
                
                seen.add(s[j_number])
                max_length = max(max_length, j_number - i_number + 1,)
        
        return max_length
        
Solution.lengthOfLongestSubstring("abcabcbb")
```
