Python Looping

```python
"""
For loop
"""
list_of_name = ['James', 'Michael', 'John', 'Robert', 'David']
list_of_number = [ 1, 2, 3, 4, 5 ]
list_of_mixed_data = [1, '2', (1, 2), {"name": "James"}]
    
# Normal looping 
for idx, data in enumerate(list_of_name):
    if data == 'John':
        continue
    elif data == 'Robert':
        break
    print(f"{idx} - name: {data}")

print('---------------------------')

# idx based looping range(start, end, step)
for idx in range(0, len(list_of_name), 1):
    print(f"{idx} - name: {list_of_name[idx]}")
    
print('---------------------------')

# idx based looping 
for idx, name in zip(list_of_number, list_of_name):
    print(f"{idx} - name: {name}")
    
print('---------------------------')

# pattern print
for idx in range(1, 5):
    var = f"{' ' * (4 - idx)}{'*' * idx}"
    print(var)
    
# reverse pattern print
for idx in range(4, -4, -1):
    print(idx)
```

```python
""" 
    While loop
"""

while_condition, count = True, 0
while while_condition:
    print(f"while looping count: {count}")
    if count == 3:
        while_condition = False
    count += 1
    
    

```
