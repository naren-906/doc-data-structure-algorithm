```python
import time

# todo: O(n) — Linear Time
""" based on list or input length time will increase """

def get_targeted_item(list_of_data: list, target: int) -> int:
    start_time = time.time()  # ⏱️ Start timer
    
    for individual_number in list_of_data:
        if target == individual_number:
            end_time = time.time()  # ⏱️ End timer
            print(f"Execution Time: {end_time - start_time:.10f} seconds")
            return individual_number
    
    end_time = time.time()
    print(f"Execution Time: {end_time - start_time:.10f} seconds")
    return None


# Test data
print("get_targeted_item:", get_targeted_item(range(900000), 899999))
```

```python
import time

# todo: O(log n) — binary search
""" The input size decreases by half each time — like divide & conquer """

def binary_search(list_of_data: list, target: int) -> int:
    start_time = time.time()  # ⏱️ Start timer
    
    if target not in list_of_data:
        return "targeted value not found"
    
    steps, low, high = 0, 0, len(list_of_data) - 1
    while low <= high:
        mid: int = (high + low)//2
        steps += 1
    
        if list_of_data[mid] == target:
            end_time = time.time()
            print(f"Execution Time: {end_time - start_time:.10f} seconds")
            return f"position of targeted value is {mid} and step {steps}"
    
        elif list_of_data[mid] < target:
            low = mid + 1
    
        elif list_of_data[mid] > target:
            high = mid - 1
    
    end_time = time.time()
    print(f"Execution Time: {end_time - start_time:.10f} seconds")
    return None

# Test data
print("binary_search: ", binary_search(range(5), 4))
```
