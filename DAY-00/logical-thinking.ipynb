```python
def print_star(input):
    for number in range(input):
        print('* '*input)

print_star(5)
```

```python
def print_star(input):
    for number in range(input+1):
        print('* '*number)

print_star(5)
```

```python
"""
summary: 
"""
def print_numbers(input):
    list_number = [str(idx) for idx in range(1, input+1)] 
    for number in range(1, input+1):
        print(' '.join(list_number[:number]))

print_numbers(5)
```

```python
"""
summary: 
"""
def print_numbers(input):
    for number in range(1, input+1):
        print(f'{number} '*number)

print_numbers(5)
```

```python
"""
summary: 
"""
def print_reverse_numbers(input):
    list_number = [str(idx) for idx in range(1, input+1)] 
    for number in range(input, 0, -1):
        print(' '.join(list_number[:number]) )

print_reverse_numbers(5)
```

```python
"""
summary: 
"""
def pyrmid_star(input):
    for number in range(1, input+1):
        space = input - number
        star = (2 * number) - 1
        print(' ' * space + '*' * star)
        
pyrmid_star(5)
```

```python
"""
summary: 
"""
def reversed_pyramid_star(input):
    for number in range(input, 0, -1):
        space = input - number
        star = (2 * number) - 1
        print(' ' * space + '*' * star)
        
reversed_pyramid_star(6)
```

```python
pyrmid_star(5)
reversed_pyramid_star(5)
```
