```python
"""
    summary: Counter Module
    description: 
        - Counter is a subclass of dict that helps count hashable objects.
        - It is an unordered collection where elements are stored as dictionary keys and their counts are stored as dictionary values.
        - It is particularly useful for counting occurrences of items in an iterable, such as a list or a string.
    
    example:
        - Counting occurrences of elements in a list.

    my _experience:
        - if u wanna find same elements in a list or iterable, Counter is very useful.
            eg: {apple: 3, banana: 2, cherry: 1}
"""

print("List of use case:")
print(" - Restaurant - Combining orders from multiple tables")
print(" - E-commerce - Merging shopping carts")
print(" - Gaming - Combining player scores from different levels")
print(" - Warehouse - Items shipped vs items in stock")
print(" - Library - Books borrowed vs available")
print(" - Event Management - Tickets sold vs refunded")
print(" - Server Monitoring - Peak concurrent users")
print(" - Retail - Maximum stock needed across stores")
print(" - Cloud Storage - Maximum file versions kept")
print(" - Supply Chain - Common minimum stock in warehouses")
print(" - Subscription Service - Minimum concurrent streams")
print(" - Manufacturing - Minimum production capacity")
print(" - Multi-Store Inventory Management System")
print("")

list_of_fruits = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
fruit_counter = Counter(list_of_fruits)
print("letter_counter: ", fruit_counter)
print("")

print("Name: Narenthiran")
letter_counter = Counter('Narenthiran')
print("     letter_counter: ", letter_counter)
print("     letter_counter: ", letter_counter.total())
print("")

scores = Counter(teamA=0, teamB=0)
scores['teamA'] += 3
print("Scores: ", scores)

errors = Counter(syntax=5, runtime=3, logical=1)
print("errors: ", errors)
print("")


# ! Built-in Function
print("Built-in Function:")
print("----------------------------")

# todo: 1) most_common() -> getting top most element from hashmap
print("Function name: counter.most_common()")
sales = Counter(['laptop', 'mouse', 'laptop', 'keyboard', 'mouse', 'laptop', 'headset'])  # Use Case: E-commerce - Top selling products
hashtags = Counter(['#python', '#ai', '#python', '#ml', '#ai', '#python']) # Use Case: Social Media - Trending hashtags

# Use Case: Social Media - Trending hashtags
print("     most_common(sales): ", sales.most_common(2))        # [('laptop', 3), ('mouse', 2)]
print("     most_common(trending): ", hashtags.most_common(3))  # [('#python', 3), ('#ai', 2)]
print("")

# todo: 2) element() -> iterate elements and making duplicate value list * value 
print("Function name: counter.elements()")
votes = Counter({"Song A": 3, "Song B": 1, "Song C": 0})
inventory = Counter({"apple": 3, "banana": 4})

print("     elements(votes): ", list(votes.elements()))
print("     elements(inventory): ", list(inventory.elements()))
print("")

# todo: 3) update() -> adding new name with existing counter hashmap
print("Function name: counter.update()")
store1_sales = Counter(["laptop", "mouse", "keyboard"])
store2_sales = Counter(["laptop", "mouse", "keyboard"])

print("     update-sales1-before: ", store1_sales)
store1_sales.update(store2_sales)
print("     update-sales1-after: ", store1_sales)

survey = Counter({"yes": 10, "no": 5})
survey.update(['yes', 'yes', 'maybe'])
print("     update-survey: ", survey)
print("")

# todo: 4) subtract() -> subtract value from existing
print("Function name: counter.subtract()")
# Use Case: Inventory Management
stock = Counter(apple=50, banana=30, orange=20)
sold = Counter({"apple": 10, "banana": 5})

print("     subtract-before(inventory): ", stock)
stock.subtract(sold)
print("     subtract-after(inventory): ", stock)

# Use Case: Budget Tracking
budget = Counter({'food': 500, 'transport': 200})
expense = Counter({'food': 150, 'transport': 50})
budget.subtract(expense)
print("     subtract(budget): ", budget)
print("")

# todo: 5) total() -> sum of all count
print("Function name: counter.total()")
# Use Case: Total votes in election
votes = Counter(candidate_a=1500, candidate_b=1200, candidate_c= 800)
print("     total(votes-count): ", votes.total())
print("")


# todo: 6) Mathematical Operations -,+,|,&
print("Function name: Mathematical Operations -,+,|,&")
# Use Case: Merging customer orders
order1 = Counter({'pizza': 2, 'burger': 1})
order2 = Counter({'pizza': 1, 'salad': 2})

NR_order_01 = { 'apple': 6, 'banana': 2, 'cherry': 1 }
NR_order_02 = { 'apple': 3, 'banana': 4, 'date': 5 }

add_total_order = order1 + order2
print("     addition(+): ", add_total_order)  # Counter({'pizza': 3, 'salad': 2, 'burger': 1})

sub_total_order = order1 - order2
print("     subtract(-): ", sub_total_order)  # Counter({'pizza': 1, 'burger': 1})

union_total_order = order1 | order2
nr_union_order = NR_order_01 | NR_order_02
print("     union(|): ", union_total_order)  # Counter({'pizza': 2, 'salad': 2, 'burger': 1})
print("     union-nr(|): ", nr_union_order)  # {'apple': 6, 'banana': 4, 'cherry': 1, 'date': 5}

inter_total_order = order1 & order2
print("     intersection(&): ", inter_total_order)  # Counter({'pizza': 1})
print("")

# todo: 7) Practical use case: Website Analytics Dashboard
print("Use Case: Website Analytics Dashboard")
page_views = Counter()

#  Day 1 traffic
page_views.update(['home', 'about', 'home', 'contact', 'products', 'home'])
print("     Day 1 Traffic: ", page_views)

#  Day 2 traffic
page_views.update(['home', 'products', 'about', 'home'])
print("     Day 2 Traffic: ", page_views)

# Top 3 pages
top_pages = page_views.most_common(2)
print("     Top 3 most visited pages:", top_pages)
print("")


# todo: 8) Advanced use case: Word Frequency Analysis
print("Use Case: Word Frequency Analysis")
artical_01 = Counter("python is interpeter language".split())
artical_02 = Counter("c++ is a compiler language".split())

# Find common trending words
common_words = artical_01 & artical_02
print("     Common Words: ", artical_01)

# Combined keyword frequency
all_keywords = artical_01 + artical_02
print("     Top Word: ", all_keywords.most_common(2))
print("")

# todo: 8) Use Case: Multi-Store Inventory Management System
print("Use Case: Multi-Store Inventory Management System")
"""
    i have three store - NY, LA, TX
    each has iphone, samsung, google phones

    task: 
        - store: get total inventory from all the location
        - sales: update or add sales based on location
        - remining product by location
        - confirm availability combined all store
"""
# Store inventories
store_NY = Counter({'Iphone': 50, 'Samsung': 28, 'Google': 40})
store_LA = Counter({'Iphone': 30, 'Samsung': 30, 'Google': 10})
store_TX = Counter({'Iphone': 25, 'Samsung': 29, 'Google': 45})

total_product = store_NY + store_LA + store_TX
print("     Total Product:", total_product)

sales_NY = Counter({"Iphone": 10, "Samsung": 5, "Google": 4})
sales_LA = Counter({"Iphone": 4, "Samsung": 2, "Google": 1})

remaining_NY = store_NY - sales_NY
remaining_LA = store_LA - sales_LA
print(f"     remaining_NY : {remaining_NY}, remaining_LA: {remaining_LA}")


```

```python
"""
    summary: defaultdict Module
    description:
        - 
    use_case:
        - Grouping values by keys from a list of key-value pairs.
        - Invert dictionary where multiple keys map to the same value.
            example: invert {'a': 1, 'b': 2, 'c': 1} to {1: ['a', 'c'], 2: ['b']}
        - Count frequencies of items
        - Group dictionary values for processing
        - Matrix or nested dictionary
    
    my_experience:
        - Learned about it will take callable function as default factory.
        - You can group a value to key very easy and quickly without checking key error.
"""

list_of_pairs = [('a', 1), ('b', 2), ('a', 3), ('c', 4), ('b', 5)]
default_dict = defaultdict(list)
for key, value in list_of_pairs:
    default_dict[key].append(value)

print("defaultdict content: ", default_dict)  # defaultdict(<class 'list'>, {'a': [1, 3], 'b': [2, 5], 'c': [4]})


# Matrix or nested dictionary
marks = defaultdict(lambda: defaultdict(int))
"""
why lambda mentioned here? 
        -> Inside defaultdict only accects callable function like list, int, str, etc or user defined function.
        -> if we dont have a value for a key then it will create a new defaultdict(int) as value.
        -> so that we can directly assign values to inner dictionary without worrying about KeyError.
"""

marks["Ram"]["Math"] = 90
marks["Ram"]["English"] = 85
marks["Sam"]["Science"] = 89

print("marks content: ", dict(marks))
# {'Ram': {'Math': 90, 'English': 85}, 'Sam': {'Science': 89}}
```
